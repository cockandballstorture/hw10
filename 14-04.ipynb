```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def func(x):
    return np.sin(x**5)+np.cos(x**2)+x**3+4*x
def der(x):
    return 5*(x**4)*np.cos((x**5))-2*x*np.sin(x**2)+3*(x**2)+4
eps=0.001
x1=1
```

```python
x = range(-5,15)
y = [func(xn) for xn in x]
plt.plot(x, y)
plt.show()
```

```python
x=0
y=func(x)
w = {x: y}
eps=0.001
for i in range(50):
    x=x*x-der(x)*eps
    y=func(x)
    w[x]=y
```

```python
plt.plot(list(w.keys()), list(w.values()))
plt.show()
q = {}
for i in range(len(w)):
    q[list(w.values())[i]] = list(w.keys())[i]
print ('мин. Х =', min(q.items())[1])
print ('мин. Y =', min(q.items())[0])
```

```python
def data_generator(numElems):
    data = np.empty((sum(numElems), 2))
    for cl,n in zip(list(range(numElems))):
        start_pos = sum(numElems[:cl])
        x_c, y_c = random.random()*5, random.random()*5
        data[start_pos:start_pos+n,:] = np.random.normal([x_c,y_c], 0.5, (n, 2))
        
    return data
```

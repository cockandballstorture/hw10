```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
figsize=(10,10)
a=pd.read_csv("advertising.csv")
b=a["TV"]
c=a["Radio"]
d=a["Newspaper"]
e=a["Sales"]
b=np.array(b)
c=np.array(c)
d=np.array(d)
e=np.array(e)
plt.figure(figsize=(10,10))
plt.subplot(1,4,1)
plt.plot(b)
plt.title('TV')
plt.subplot(1, 4,2)
plt.plot(c)
plt.title('Radio')
plt.subplot(1, 4,3)
plt.plot(d)
plt.title('Newspaper')
plt.subplot(1, 4,4)
plt.plot(e)
plt.title('Sales')
```

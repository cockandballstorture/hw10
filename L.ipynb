```python
class MSE:
    def compute(self, y, y_pred):
        return ((y_pred-y)**2).sum()/y.shape[0]
class LinearRegression:
    def __init__(self, x,y):
        self.x=x
        self.y=y
    def fit(self, x, y):
        xmean=x.mean()
        dx=xmean-x
        dsx=dx**2
        sx=dsx.sum()
        ymean=y.mean()
        dy=ymean-y
        dsy=dx**2
        sy=(dx*dy).sum()
        self.a=sy/sx
        self.b=ymean-self.a*xmean
    def predict(self, x):
        return self.a*x+self.b
model = LinearRegression(x,y)
model.fit(x,y)
pred=model.predict(112)
pred
```
